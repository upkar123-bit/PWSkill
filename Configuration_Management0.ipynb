**Configuration Management**
1. Research and compare different configuration management tools (Ansible,Puppet, Chef) and write a report on their use cases and advantages?

**Ansible**

Architecture: Control node pushes changes over SSH to target nodes (agentless). Inventory + playbooks (YAML).


Language/style: YAML playbooks + modules; mostly imperative tasks in sequence (but you can write declarative roles).


Ecosystem: Red Hat Ansible Automation Platform commercial offering adds UI, policy, RBAC and automation controller.


**Puppet**

Architecture: Agent-server model (Puppet agent runs on nodes and polls the Puppet server / master) though Puppet also supports other modes (apply). Strong model-driven approach.



Language/style: Puppet DSL (declarative). You declare desired state; Puppet enforces it and reports drift.


**Chef**

Architecture: Client-server (Chef Server + Chef Client) plus Chef Workstation. Also supports local mode and Chef Solo.


Language/style: Ruby-based DSL (recipes and cookbooks). More programmatic; good for conditional logic and reusable libraries.

When to choose Ansible
* You want fast setup and a low learning curve (YAML). Good for teams without heavy programming background.

When to choose Puppet
* You need declarative, model-driven configuration for consistent state enforcement across thousands of nodes. Puppet's reporting and compliance features are strong for enterprises

When to choose Chef
* You need highly programmatic configuration (complex conditional logic, libraries) and strong test-driven workflows (ChefSpec, Test Kitchen). Chef is favored where configuration is complex and code-centric practices are used.

1. Advantages

 * Ansible
 1. Agentless (less maintenance).
 2. Simple YAML syntax → quick onboarding.
 3. Great for orchestration beyond configuration.

* Puppet
1. Declarative, model-driven approach — good for compliance and drift enforcement at scale.
2. Mature reporting, role/class system and enterprise support (Puppet Enterprise).

* Chef
1. Flexible programmatic DSL for complex logic, strong testing ecosystem.
2. Good for organizations that treat infrastructure as code with software engineering practices.

Q2. Extend the CI pipeline to include continuous delivery using Jenkins or
GitLab CI. Implement a blue-green deployment strategy?

A standard CI pipeline handles code building, testing, and artifact generation. To achieve full automation, organizations extend CI into Continuous Delivery (CD), where each successful build is automatically deployed to a staging or production environment.

Using tools like Jenkins or GitLab CI, we can automate the entire lifecycle:
→ Build → Test → Package → Deploy → Switch traffic (Blue-Green)


**Continuous Delivery ** ensures that every change that passes tests is ready to deploy automatically. Key features include:

Automated deployment to staging/production

Versioned artifacts (Docker images, packages)

Automated configuration and environment provisioning

Safe rollout strategies (e.g., Blue-Green, Canary)

1.  Extending CI into CD Using Jenkins
 * Jenkins Workflow
1. Developer pushes code = GitHub/GitLab
2. Jenkins triggers CI pipeline
3. Build + Unit tests
4. Create Docker image
5. Push image to Docker Registry
6. CD stage: Deploy updated version to Green environment
7. Run smoke tests
8. Switch traffic from Blue to Green

```python
example Jenkinsfile (Conceptual)
pipeline {
    agent any

    stages {
        stage('Build') {
            steps {
                sh 'mvn clean package'
            }
        }

        stage('Docker Build & Push') {
            steps {
                sh 'docker build -t myapp:${BUILD_NUMBER} .'
                sh 'docker push myapp:${BUILD_NUMBER}'
            }
        }

        stage('Deploy to Green') {
            steps {
                sh './deploy-green.sh ${BUILD_NUMBER}'
            }
        }

        stage('Smoke Test Green') {
            steps {
                sh './smoke-test-green.sh'
            }
        }

        stage('Switch Traffic to Green') {
            steps {
                sh './switch-blue-green.sh'
            }
        }
    }
}
```
