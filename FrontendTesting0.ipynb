Load testing evaluates how a web application behaves when multiple users access it simultaneously. Tools like Apache JMeter and Gatling simulate virtual users, measure response times, and detect bottlenecks before deployment.

**Load testing helps determine**:
* Maximum requests the system can handle
* Response time under load
* Application stability
* How the system scales with increasing users

**Apache JMeter**

A Java-based performance testing tool ideal for:
* HTTP/HTTPS testing
* APIs
* Databases
* Distributed load tests

**Key Features**
* GUI and CLI mode
* Pluggable samplers
* CSV reporting
* Supports correlation & parameterization

Analyzing Load Test Results
1. High Response Time
2. High Error Rate
3. Low Throughput
4. Increased Response Time at Higher Load

**Based on test data, bottlenecks can be found in:**

A. Application Layer
B. Database Layer
C. Server Hardware
