Theoretical Overview of Docker and Containerization
1. Install Docker and create Dockerfiles to containerize applications?

Docker is an open-source platform that allows you to build, package, and run applications inside containers.
A container is a lightweight, isolated, and portable environment that includes:

Your application code

Required libraries

Dependencies

Runtime environment

This ensures that the application runs the same on any system—developer laptop, testing environment, or production server.

**Containerization**

Containerization is the process of packaging software and its dependencies into a single unit called a container.

Benefits:

Portability: Works anywhere (Linux, Windows, Cloud).

Lightweight: Uses fewer resources compared to VMs.

Isolation: Each container runs independently.

Faster Deployment: Start/stop in seconds.

Consistency: "Works on my machine" issues disappear.

**To use Docker, you install Docker Engine, which includes:**

Docker daemon → Runs containers.

Docker CLI → Command-line tool to interact with Docker.

Docker Compose (optional) → Runs multi-container applications.

**Dockerfile**

A Dockerfile is a text file containing instructions that tell Docker how to build an image.

```python
docker build -t app-image .
```

**How Containerization Works (Theoretical Flow)**

**Step 1**: Write a Dockerfile

You define:
* Base OS
* Runtime (Node, Python, Java, etc.)
* App dependencies
* Commands to run the app

**Step 2**: Build a Docker Image

* Docker reads the Dockerfile and packages everything into an image.

**Step 3**:
* Run a Container
* You create a container from the image:
* docker run -d -p 3000:3000 my-app


Now your application runs in an isolated environment.
